```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris['data'])
df['category'] = iris.target
df.columns = ['sepal_length','sepal_widht','petal_length','petal_width','Category']
```

# Whole Table

```python
# analysis of whole table 
df.describe()
```

```python
# displaying the categories
iris.target_names
```
